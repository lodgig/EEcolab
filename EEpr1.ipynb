# 1 генерация




```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
X, y_blob= make_blobs(n_samples =100, centers=3, n_features=2, random_state=0, cluster_std=0.06, center_box=(0,1))
X[:,0] = ((X[:,0]-0.01)/0.99)*300+1
#X[:,1]  = X[:,1]  
```

```python
#X[:,0]  каждый 1
#X[:,1]  каждый 2
```

```python
plt.scatter(X[:,0],X[:,1])
```

# 2 кластеризация


```python
from sklearn.cluster import KMeans
kmeans1 = KMeans(n_clusters = 3)
kmeans1.fit(X)
```

```python
y_kmeans = kmeans1.predict(X)
data = np.array([X[:,0],X[:,1],y_kmeans]).transpose()
pd.DataFrame(data, columns=['X1', 'X2', 'cluster'])
```

# 3 визуализация


```python
centers = kmeans1.cluster_centers_
plt.scatter(X[:,0],X[:,1],c=y_kmeans, cmap='viridis')
plt.scatter(centers[:,0],centers[:,1],c="red")
```
